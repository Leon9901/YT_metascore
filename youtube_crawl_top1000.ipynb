```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
#抓第一頁
res = requests.get('https://www.channelcrawler.com/eng/results/26787/sort:Channel.subscribers/direction:desc')
res.text
```

```python
soup = BeautifulSoup(res.text, 'html5lib')
```

```python
item = soup.find_all(href = re.compile('www.youtube.com'))
```

```python
#萃取Channel_Name + ID
g=0
yid = []
yname = []
#for temp in soup.find_all(href= re.compile('www.youtube.com')):
for temp in item:
    g+=1
    if g%3==1:
        yid.append(str(temp)[40:64])
        yname.append(temp.get('title'))
    else:
        continue
```

```python
ytid = ""
```

```python
#抓第2-末頁
total_page = 51
for p in range(2,total_page):
    url_link = 'https://www.channelcrawler.com/eng/results/26787/page:'+ str(p) +'/sort:Channel.subscribers/direction:desc'
    resn = requests.get(url_link)
    soup = BeautifulSoup(resn.text, 'html5lib')
    item = soup.find_all(href = re.compile('www.youtube.com'))
#     for i in range(0,len(item),3):
#         dtset.append(str(item[i])[40:64])
    #for temp in soup.find_all(href= re.compile('www.youtube.com')):
    for temp in item:
        g+=1
        if g%3==1:
            yid.append(str(temp)[40:64])
            yname.append(temp.get('title'))
        else:
            continue
```

```python
ytid = pd.DataFrame({
    'ID':yid,
    'NAME':yname
})
```

```python
ytid.to_csv('ytid.csv', sep=',', encoding='utf-8', mode='a')
```
